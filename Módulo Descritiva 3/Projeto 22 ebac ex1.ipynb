#### 1. Baixando dados de ativos
    1. Baixe os dados de três ativos da sua escolha do Yahoo Finance, correspondente a um período de pelo menos 2 anos até a data mais recente.
    2. Salve esses dados para que você possa utilizá-los na próxima tarefa.

```python
import pandas as pd
import numpy as np
import plotly.express as px
from plotly import graph_objects
from pandas_datareader import data as web
import yfinance as yf
import nbformat
```

```python
# 1 Importando os dados
ativos = ['^DJI', 'DIS', 'JNJ'] #Os ativos correspondem a Dow Jones, Disney e Johnson & Johnson, respectivamente 
dt_start, dt_end = '2019-01-01', '2024-03-01'
yf.pdr_override()
df = web.DataReader(ativos,
                    dt_start,
                    dt_end)
df.head()
```

```python
df.to_csv('valores_bolsa.csv', index= False)
```

#### 2. Observando a volatilidade dentro do dia
1. Calcule a amplitude de variação do valor do ativo dentro do dia. Essa amplitude você pode calcular com um comando simples de adicionar uma coluna no Pandas: (valor máximo - valor mínimo).
2. Monte um gráfico de linha mostrando a variação dessa amplitude ao longo do período coletado, por ativo.
3. Veja se consegue identificar que ativo apresentou maior volatilidade.

```python
# 2 Primeiramente iremos montar um gráfico alternativo, que contenha somente as informações relevantes
data_graph = df[['High','Low']].stack()
data_graph['Amplitude']= data_graph['High']-data_graph['Low']
data_graph
```

```python
# Agora criaremos variáveis com os valores das datas e das ações
datas = data_graph.index.get_level_values('Date')
tickers = data_graph.index.get_level_values('Ticker')
```

```python
# Por fim, plotaremos o gráfico
fig = px.line(data_graph, x= datas, y= 'Amplitude', color= tickers, template= 'simple_white')
fig.show()
#OBS: Os valores da Dow Jones são tão grandes que os outros aparecem achatados, deselecionar estes garante melhor visualização dos outros ativos.
```
