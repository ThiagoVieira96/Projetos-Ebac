#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
#Aqui será feita uma função para avaliar o APGAR5 e criaremos depois uma nova coluna, de maneira semelhante ao exercicio 1 do módulo 5
def avaliar_apgar(col):
    if 0 < col <=3:
        return 'asfixia grave'
    elif col <=5:
        return 'asfixia moderada'
    elif col <=7:
        return 'asfixia leve'
    else:
        return 'normal'
    
    
sinasc['avaliacao_apgar5'] = sinasc['APGAR5'].apply(avaliar_apgar)
sinasc.head()
```

```python
sinasc.columns
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
fig, ax = plt.subplots(1)
ax.tick_params(axis='x', rotation=20)
sns.barplot(x = 'GESTACAO', y= 'APGAR5', hue='SEXO', data= sinasc)
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
#Esse aqui ta errado
tabela =pd.crosstab(sinasc['APGAR5'], sinasc['GESTACAO'])
tabela
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
tabela_pct = tabela.div(tabela.sum(axis=1), axis=0)
tabela_pct
```

```python
#Mostrando que a soma das linhas é 100%
tabela_pct.sum(axis=1)
```

```python
tabela_pct.plot.bar()
```

Dado que, no gráfico acima, grande parte dos valores onde o APGAR5 é menor que 6, isso é, onde o bebê teve algum tipo de asfixia que mais grave que uma asfixia leve, são de crianção que nasceram com menos de 37 semanas de gestação, e também, se observarmos que a porcentagem de crianças com menos de 37 semanas de gestação diminui conforme os valores de APGAR5 aumenta, é bastante plausível a afirmação de que tais crianças são prematuras.
