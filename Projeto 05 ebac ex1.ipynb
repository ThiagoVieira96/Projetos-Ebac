````python
# 8) seu código aqui
#Aqui será feita uma função para avaliar o APGAR5 e criaremos depois uma nova coluna
def avaliar_apgar(col):
    if 0 < col <=3:
        return 'asfixia severa'
    elif col <=5:
        return 'asfixia moderada'
    elif col <=7:
        return 'asfixia leve'
    else:
        return 'normal'
    
    
df_sinasc['avaliacao_apgar5'] = df_sinasc['APGAR5'].apply(avaliar_apgar)
df_sinasc
````
